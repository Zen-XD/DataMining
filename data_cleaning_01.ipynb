```python
#@ Import dataset using pandas library
import pandas as pd
df=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Lab 1/Customer Call List.xlsx')
df.head(20)
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns='Not_Useful_Column')
df
```

```python
#@ Removing unnecessary chars from Last Name
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace('[^0-9]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x if len(str(x)) == 10 else '')
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df
```

```python
#@ Parse Address columns into Street_Address, State and Zip_code
df[['Street_Address', 'State', 'Zip_code']] = df['Address'].str.split(',', expand=True)
df = df.drop(columns='Address')
df
```

```python
#@ lets make Paying costumer to be Y for Yes and N for No
df['Paying Customer'] = df['Paying Customer'].replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
#@ lets make Paying costumer to be Y for Yes and N for No
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
df = df.replace(['N/a', None], '')
df = df.fillna('')
df
```

```python
df = df[df['Do_Not_Contact'] != 'Y']
df
```

```python
#@ If there is no phone number remove that row
df = df[df['Phone_Number'] != '']
df
```

```python
df = df.reset_index(drop=True)
df
```
